```python
import pandas as pd
import numpy as np

data = {
    "OrderID": [101,102,103,104,105,106,107,108,109,110],
    "CustomerID": ["C1","C2","C3","C4","C5","C6","C7","C8","C9","C10"],
    "ProductCategory": ["Electronics","Electronics","Electronics","Electronics",
                        "Electronics","Clothing","Clothing","Furniture","Furniture","Electronics"],
    "SalesAmount": [25000, np.nan, 30000, np.nan, 28000, 1500, 2000, 8000, 9000, np.nan],
    "Region": ["South","South","South","South","South","North","South","West","West","East"],
    "OrderDate": pd.date_range(start="2025-01-01", periods=10)
}

df = pd.DataFrame(data)

df["Imputation_Method"] = "None"

regional_median = df.groupby(["ProductCategory","Region"])["SalesAmount"].median()

missing_index = df[df["SalesAmount"].isnull()].index

for i in missing_index:
    category = df.loc[i, "ProductCategory"]
    region = df.loc[i, "Region"]
    median_value = regional_median.get((category, region))
    
    if not pd.isna(median_value):
        df.loc[i, "SalesAmount"] = median_value
        df.loc[i, "Imputation_Method"] = "Regional_Median"

category_median = df.groupby("ProductCategory")["SalesAmount"].median()

for i in df[df["SalesAmount"].isnull()].index:
    category = df.loc[i, "ProductCategory"]
    df.loc[i, "SalesAmount"] = category_median[category]
    df.loc[i, "Imputation_Method"] = "Category_Median"

df.to_csv("cleaned_retail_sales.csv", index=False)

print(df)
print(df["Imputation_Method"].value_counts())
```
