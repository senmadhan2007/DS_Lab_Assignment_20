```python
import pandas as pd

data = {
    "Employee_Name": [
        "Dr. Rahul Kumar Sharma","Priya M.","Mr. S. Venkatesh","Ms. Anjali Gupta",
        "Rahul Sharma","Dr. Meena S Iyer","Mr. K. Ramesh","Priya M.",
        "Ms. Kavya Nair","Dr. Arjun Reddy","Mr. S. Venkatesh","Anil Kumar",
        "Dr. Rahul Kumar Sharma","Ms. Sneha Kapoor","Mr. Rohit Sharma",
        "Kiran Rao","Dr. Pooja Singh","Ms. Anjali Gupta","Mr. Suresh Babu",
        "Meena Iyer","Dr. Arjun Reddy","Ms. Kavya Nair","Rohit Sharma",
        "Dr. Sneha Kapoor","Mr. K. Ramesh"
    ]
}

df = pd.DataFrame(data)

df["Employee_Name"] = df["Employee_Name"].str.replace(r"(Dr\.|Mr\.|Ms\.)","", regex=True).str.strip()

first = []
middle = []
last = []

for name in df["Employee_Name"]:
    parts = name.split()
    if len(parts) == 3:
        first.append(parts[0])
        middle.append(parts[1][0])
        last.append(parts[2])
    elif len(parts) == 2:
        first.append(parts[0])
        middle.append("")
        last.append(parts[1])
    else:
        first.append(parts[0])
        middle.append("")
        last.append("")

df["First_Name"] = first
df["Middle_Initial"] = middle
df["Last_Name"] = last

usernames = []
counts = {}
duplicates = 0

for i in range(len(df)):
    base = (df.loc[i,"First_Name"][0] + df.loc[i,"Last_Name"]).lower()
    if base in counts:
        counts[base] += 1
        username = base + str(counts[base])
        duplicates += 1
    else:
        counts[base] = 1
        username = base
    usernames.append(username)

df["Username"] = usernames

print(df.head(8))
print("Duplicate usernames handled:", duplicates)
```
